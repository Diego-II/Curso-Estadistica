### 2.B Cargar datos

Numpy permite cargar archivos en formato txt, csv.

_Observación:_ El formato 'csv' _(comma-separated values)_ es un archivo cuyos datos estan separados por comas.

**Nota**: Tambien se puede usar [Pandas](https://pandas.pydata.org/) _(Python Data Analysis Library)_ , una librería especializada en el manejo de archivos csv, excel y HDF5 entre otros. En particular, Pandas es útil para bases de datos más complejas y no necesariamente compuestas de solo números.

> Antes de cargar nuestros datos, creemos unos datos artificiales y guardémoslos, donde más adelante los cargaremos como si fuese un archivo externo.
 Para crear nuestros datos usemos NumPy arrays. 
> En este caso obtendremos muestras de una Gaussiana multivariada en 2 dimensiones, para esto:
1. Definimos un vector de medias de tamaño 2
2. Definimos una matríz de covarianza de 2x2 
3. Obtenemos 50 muestras de esta distribución usando NumPy

````python
# media = [3,1]
mu = np.array([3, 1])
# matriz covarianza
cov = np.array([[1, 0.5], [0.5, 1]])
# cantidad de muestras
n = 50

# obtenemos las muestras (observaciones)
samples = np.random.multivariate_normal(mean=mu, cov=cov, size=(50))
print('Las muestras generadas son de tamaño', samples.shape)
````

> Ahora procedemos a guardar nuestros datos.

````python
nombre_archivo = 'datos.csv'
# función para guardar archivo
np.savetxt(nombre_archivo, X=samples, delimiter=',')
````

> Con los datos guardados, podemos cargarlos y empezar a trabajar con ellos.


> __Primero__: dividamos las muestras según eje, donde las primera dimensión la llamaremos $X$ y la segunda $Y$. Luego utilizemos NumPy para obtener intuición inicial de los datos.

````python
datos = np.loadtxt('datos.csv', delimiter=',')
print('Datos cargados tienen tamaño', datos.shape)
````

````python
# asignar muestras en X
x = datos[:, 0]
# asignar muestas del eje Y
y = datos[:, 1]

print('El arreglo en el eje x es de tamaño', x.shape)
````

> Obtengamos la media de cada eje y veamos que tan cerca está del valor real.

````python
# obtener media
media_x = x.mean()
print('Media empírica en X', media_x)

# media en y la calculamos de forma directa en el print
print('Media empírica en Y', y.mean())
````

### 4.B Optimización

> Utilizemos SciPy para optimizar una función. Para esto primero crearemos una función de varios parámetros y luego minimizemos respecto a un subconjunto de éstos.

> Definamos nuestra función a optimizar, en este caso será de la forma, 
$$ J = \alpha \left[(x - \beta)^2 + (y - \gamma)^2 \right]$$
Donde $\alpha>0, \beta, \gamma$ son parámetros __fijos__ del a función, donde minimizaremos respecto a $x$ e $y$.

> Para distinguir entre los parámetros fijos de nuestra función, el primer argumento será un ndarray dado por $z=(x,y)$, y los siguiente argumentos serán los demás parámetros.
Dado la forma de la función sabemos que el mínimo está en $(x, y)^* = (\beta, \gamma)$ con $J[(x, y)^*]=0$.

````python
#definimos la función. El primer argumento DEBE ser la variable de optimización
def funcion_obj(z, alpha, beta, gamma):
    """
    Función objetivo a optimizar.
    """
    x = z[0]
    y = z[1]
    
    J = (x - beta)**2 + (y - gamma)**2
    J = J * alpha
    return J

# probemos la función
a = 0.5 # alpha
b = 20  # beta
c = -10 # gamma

x = 543
y = 500
z = np.array([x, y])
print('Valor función objetivo es', funcion_obj(z, a, b, c))
````

````python
# importemos el optimizador
from scipy.optimize import minimize
````

> **Ahora optimizemos**

````python
# punto inicial al optimizar
z_inicial = np.array([500, 500])

# opciones extras
opciones = {'maxiter':1000, 'disp':True}

res = minimize(
        fun=funcion_obj,          # funcion objetivo
        x0=z_inicial,             # punto inicial
        args=(a, b, c),           # argumentos fijos
        method='L-BFGS-B',        # metodo optimización: Low memory Broyden–Fletcher–Goldfarb–Shanno con restricciones
        options=opciones          # argumentos extras
        )
````

> Ahora que tenemos nuestro resultado podemos obtener que valor obtuvo el optimizador

````python
#óptimo real
print('Valor de (x,y) optimo real:', [b,c])

# valor de z=(x,y) optimo
z_optimo = res.x
print('Valor de (x,y) optimo encontrado:', z_optimo)

# valor de la funcion objetivo en el optimo
J_optimo = res.fun
print('Valor de funcion objetivo minima encontrada:', J_optimo)
````

> Por último, intentemos graficar esta función con un mapa de colores, para esto hagamos una malla de valores para X e Y y luego grafiquemos.

````python
# crear equiespaciado para evaluar función
x = np.linspace(-50, 50, 100)
y = np.linspace(-50, 50, 100)

# crear malla
xx, yy = np.meshgrid(x, y)
# evaluar malla
jj = funcion_obj([xx, yy], a, b, c)
````

> Ahora grafiquemos la función, otra cosa importante que haremos es guardar la figura en graficos vectorizados para evitar perder resolución.

````python
plt.figure(figsize=(8, 6))
# graficar contorno
plt.contour(xx, yy, jj, 120, cmap='RdGy')
plt.plot(z_optimo[0], z_optimo[1], marker='*', markersize=30, color='g', label='Óptimo encontrado')
plt.plot(b, c, marker='.', markersize=15, color='r', label='Óptimo real')
# muestra legenda
plt.legend()

# fija nombre de ejes
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Mapa de calor y minimización usado BFGS')

# muestra legenda de color
plt.colorbar()

# elimina espacios inncesarios en figura
plt.tight_layout()

plt.savefig('figura.pdf')
````
